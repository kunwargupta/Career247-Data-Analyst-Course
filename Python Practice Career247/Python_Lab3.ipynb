**Exercise 1:** Wrap a division `a/b` in try/except to catch `ZeroDivisionError` and print a helpful message.

```python
a = 5
b = 0

try:
    print(a/b)
except ZeroDivisionError:
    print("Do not divide by zero")
```

**Exercise 2:** Convert the string `'42'` to int inside try/except/else/finally; print success in else, 'done' in finally.

```python
try:
    a = int('abc')
except:
    print("wrong value in int()")
else:
    print('success')
finally:
    print('done')
```

**Exercise 6:** Define a custom exception class `TooSmallError`. Write a function that raises it if input < 10; test with 7 and handle it.

```python
class TooSmallError(Exception):
    pass

def check(num):
    if num < 10:
        raise TooSmallError(f"{num} is too small! Need 10 or more.")
        
try:
    check(7)
except TooSmallError as e:
    print("error:", e)
```

# Standard Library Focus: math, random, datetime (Heavier)

**Exercise 1:** Using `math`, compute area of a circle (r=7) and print to 3 decimals.

```python
import math

r = 7
res = math.pi * r**2
print(round(res,3))
```

**Exercise 2:** Using `math`, compute factorial of 6 and the gcd of 84 and 30; print both.

```python
n = 6
res1 = math.factorial(n)

n1 = 84
n2 = 30
res2 = math.gcd(n1,n2)
print(f"Factorial of {n} is {res1}, GCD of {n1},{n2} is {res2}")
```

**Exercise 3:** Compute compound interest for P=1500, r=6% annual, t=4 years (annual compounding). Print rounded to 2 decimals.

```python
p = 1500
r = 6
t = 4
a = p * (1+r/100) ** t
print(round(a,2))
```

**Exercise 4:** Generate a reproducible list of 10 random integers from 1..100 using a fixed seed (123).

```python
import random
random.seed(123)
nums = [random.randint(1,100) for _ in range(10)]

print(nums)
```

**Exercise 5:** From a list `['red','green','blue','yellow']`, randomly shuffle the list and print it. Then sample 2 unique colors.

```python
import random
random.seed(15)
lst = ['red', 'green', 'blue', 'yellow']
random.shuffle(lst)
print(lst)
print(random.sample(lst, 2))
```

**Exercise 6:** Simulate 1000 coin flips (`H`/`T`) and print the counts of each.

```python
coin = ['H', 'T']
flips = [random.choice(coin) for _ in range(1000)]
print(f"Heads: {flips.count('H')}")
print(f"Tails: {flips.count('T')}")
print(flips)
```

**Exercise 7:** Print today's date and time, then print the date 30 days from now.

```python
from datetime import datetime, timedelta 

today = datetime.now()
print(today)
print(f"Today: {today.strftime("%d-%m-%y %H:%M:%S")}")

future = today + timedelta(days=30)
print(f"30 days from now: {future.strftime("%d-%m-%y %H:%M:%S")}")
```

**Exercise 8:** Given ISO date strings `['2025-05-01','2024-12-31','2025-01-15']`, sort them chronologically using `datetime`.

```python
dates = ['2025-05-01','2024-12-31','2025-01-15']
sort_date = sorted(dates, key=lambda x: datetime.fromisoformat(x))
print(sort_date)
```
